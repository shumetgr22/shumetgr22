```python
import numpy as np
import matplotlib.pyplot as plt

def exponential_pdf(x, lambd=1):
    return lambd * np.exp(-lambd * x)

def inverse_exponential_cdf(u, lambd=1):
    return -np.log(1 - u) / lambd

# Generate 1000 uniform random numbers
uniform_samples = np.random.rand(1000)

# Apply Inverse CDF to get exponential samples
exponential_samples = inverse_exponential_cdf(uniform_samples)

# Plot Histogram
plt.hist(exponential_samples, bins=30, density=True, edgecolor='black')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Histogram of 1000 Samples from Exponential Distribution (Î»=1)')
plt.show()
```
