```python
import numpy as np
import matplotlib.pyplot as plt

# Given pmf
values = np.array([1, 2, 3, 4, 5, 6])
pmf = np.array([0.1, 0.3, 0.2, 0.1, 0.2, 0.1])

# Calculate CDF
cdf = np.cumsum(pmf)

# Generate 500 uniform random numbers
uniform_samples = np.random.rand(500)

# Apply Inverse Transform
inverse_transform_samples = np.searchsorted(cdf, uniform_samples)

# Plot Histogram
plt.hist(inverse_transform_samples, bins=np.arange(0.5, 7.5, 1), density=True, edgecolor='black')
plt.xticks(values)
plt.xlabel('X')
plt.ylabel('Probability')
plt.title('Histogram of 500 Samples using Inverse Transform Method')
plt.show()
```
